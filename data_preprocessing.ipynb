We split the article content based on parts since constitution is divided into 25 PARTS

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

separator = "PART "  # Common chapter heading pattern

# Initialize the RecursiveCharacterTextSplitter with your chapter separator
text_splitter = RecursiveCharacterTextSplitter(
    separators=[separator],  # List of separators used to split the document
    chunk_size=1000,         # Maximum size of each split chunk
    chunk_overlap=100          # No overlap between splits
)

# Concatenate the 'page_content' of all Document objects into a single string
text = "".join([doc.page_content for doc in articles])

# Split the combined text by chapters
chapters = text_splitter.split_text(text)
chapters = chapters[1:]
```

```python
for i in range(len(chapters)):
  print(f"part({i}) {chapters[i]}")
  print("\n\n")
```

```python
def split_article_wise(chapter):
  """ Split the text based on Articles"""
  sections = re.split(r'(?=\n\d+\[?\d*[a-zA-Z]?[a-zA-Z]?\.\s)', chapter)
  # Optional: Strip leading/trailing whitespaces from each section
  sections = [section.strip() for section in sections]
  return sections
```

```python
articles=[]
for chapter in chapters:
  articles.append(split_article_wise(chapter))
```

```python
def cleaning_articles(arti):
  """Cleaning the articles text for unwanted special characters"""
  cl_art=[]
  for i in arti:
    pattern = r'^\d+\['
    if re.match(pattern, i):
      pos=arti.index(i)
      i = i.replace(i[:2], "", 1)
      arti[pos] = i
      cl_art.append(i)
    else:
      cl_art.append(i)
  return cl_art
```

```python
docs = [cleaning_articles(i) for i in articles]
```

```python
def adding_tag(docs):
  '''For adding Article keyword before each article number'''
  d=[docs[0]]
  for i in range(1,len(docs)):
    docs[i] = "Article " + docs[i]
    d.append(docs[i])
  return d
```

# Processing IPC Dataset

```python
import pandas as pd
```

```python
ipc_df = pd.read_csv('/content/drive/MyDrive/SIC_Project/ipc_sections.csv')
print("IPC sections loaded into 'ipc_df'.")
ipc_df.head()
```

```python
from langchain_core.documents import Document

ipc_docs = []
for index, row in ipc_df.iterrows():
    # Combine relevant columns into page_content
    page_content = f"Description of {row['Section']}: {row['Description']}\nOffense: {row['Offense']}\nPunishment: {row['Punishment']}"

    # Define metadata, including existing general metadata and the specific section
    metadata = {
        "source": "/content/drive/MyDrive/SIC_Project/ipc_sections.csv",
        "source_name": "Indian Penal Code",
        "country": "India",
        "db_owner": "LawGlance",
        "section": row['Section']
    }

    ipc_docs.append(Document(page_content=page_content, metadata=metadata))

print(f"Created {len(ipc_docs)} Document objects from ipc_df.")
print("Example Document:")
print(ipc_docs[0])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the RecursiveCharacterTextSplitter
# Using similar parameters as for the Constitution to maintain consistency
t_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=300,
    length_function=len,
    is_separator_regex=False, # Changed to False as we are splitting general text now
)

# Chunk the ipc_docs
chunked_ipc_docs = t_splitter.split_documents(ipc_docs)

print(f"Original IPC documents: {len(ipc_docs)}")
print(f"Chunked IPC documents: {len(chunked_ipc_docs)}")
print("Example of a chunked IPC document:")
print(chunked_ipc_docs[0])
```

```python
output_pickle_path = "/content/drive/MyDrive/SIC_Project/ipc_section.pkl"
with open(output_pickle_path, "wb") as f:
    pickle.dump(chunked_ipc_docs, f)

print(f"Chunked IPC documents saved to {output_pickle_path}")
```

```python
from groq import Groq

# ---------------------------------------------
# Initialize Groq API
# ---------------------------------------------
client = Groq(
    api_key="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"
)

# ---------------------------------------------
# Legal Aid Assistant Function
# ---------------------------------------------
def legal_assistant(question):

    # Retrieve documents from RAG
    docs = retriever.invoke(question)

    # Convert retrieved docs to context
    context = "\n".join([doc.page_content for doc in docs])

    # If no relevant context found
    if not context.strip():
        return "Please ask relevant questions related to legal aid only."

    prompt = f"""
You are an AI Legal Aid Assistant designed to help users understand laws.

STRICT RULES:

1. Answer ONLY using the provided legal context and write in a para.
2. Don't add according to context just answer based on context.
2. If the question is NOT related to legal matters OR the context does not contain the answer,
respond ONLY with:

Please ask relevant questions related to legal aid only.

3. Do not generate information outside the context.

Legal Context:
{context}

User Question:
{question}

Answer strictly based on the context above.
"""

    response = client.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {
                "role": "system",
                "content": "You are a professional legal aid assistant."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0
    )

    return response.choices[0].message.content
```
